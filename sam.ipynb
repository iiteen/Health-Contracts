```python
import os
import shutil
```

```python
import fitz

pdf_path = "Samples\Contract 6.pdf"
pdf_document = fitz.open(pdf_path)
output_folder = "extracted_images"

# Create or clear the output folder
if os.path.exists(output_folder):
    shutil.rmtree(output_folder)  # Remove existing folder and its contents
os.makedirs(output_folder)  # Create a fresh folder

num_images_extracted = 0

for page_num in range(len(pdf_document)):
    page = pdf_document[page_num]
    images = page.get_images(full=True)
    for img_index, img in enumerate(images):
        xref = img[0]
        base_image = pdf_document.extract_image(xref)
        image_bytes = base_image["image"]

        image_filename = os.path.join(
            output_folder, f"image_{page_num}_{img_index}.png"
        )
        with open(image_filename, "wb") as image_file:
            image_file.write(image_bytes)
            num_images_extracted += 1

pdf_document.close()
print(f"Total images extracted: {num_images_extracted}")
```

```python
# pdf to image
# install pdf2image and also poppler for this to work,
# pip install pdf2image
# conda install -c conda-forge poppler
import os
from pdf2image import convert_from_path

pdf_path = "Samples\Contract 1.pdf"
images = convert_from_path(pdf_path)
output_folder = "images"

# Create or clear the output folder
if os.path.exists(output_folder):
    shutil.rmtree(output_folder)  # Remove existing folder and its contents
os.makedirs(output_folder)  # Create a fresh folder

for i, img in enumerate(images):
    img.save(os.path.join(output_folder, f"page{i}.jpg"), "JPEG")
```

```python
# pip install pytesseract
# conda install -c conda-forge tesseract
# image -> text
import pytesseract
from PIL import Image

# important step to set path
pytesseract.pytesseract.tesseract_cmd = (
    r"C:\Users\sunil\anaconda3\envs\hilabs\Library\bin\tesseract.exe"
)
os.environ["TESSDATA_PREFIX"] = r"C:\Users\sunil\anaconda3\envs\hilabs\share\tessdata"


for i in range(len(images)):
    image_path = os.path.join(output_folder, f"page{i}.jpg")
    image = Image.open(image_path)

    extracted_text = pytesseract.image_to_string(image)
    print(f"Text from page{i}:\n{extracted_text}\n\n")
```
